```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
```

```python
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
data = pd.read_csv("/content/gdrive/MyDrive/resume/dataset.csv")
```

```python
values = data["Index"].values.reshape(-1, 1)
values
```

```python
isolation_forest = IsolationForest(contamination=0.1) #assumed anomaly is 10%
```

```python
isolation_forest.fit(values)
```

```python
predictions = isolation_forest.predict(values)
```

```python
data["Anomaly"] = predictions

# Save the DataFrame with predictions to a new CSV file
output_csv_filename = "data_with_anomalies.csv"
data.to_csv(output_csv_filename, index=False)

print("Anomaly detection results saved to:", output_csv_filename)
```

```python
num_anomalies = np.sum(predictions == -1)
print("Number of Anomalies:", num_anomalies)
print("Anomaly detection results saved to:", output_csv_filename)
```

```python
true_labels = [-1] * len(data)

# Calculate metrics
accuracy = accuracy_score(true_labels, predictions)
precision = precision_score(true_labels, predictions, pos_label=-1)
f1 = f1_score(true_labels, predictions, pos_label=-1)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("F1-score:", f1)
```
